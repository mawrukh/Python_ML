```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
#convert to bs obj
page=requests.get(url)
soup=bs(page.text , 'html')
print(soup)
```

```python
table=soup.find_all('table')[0]
table
```

```python
table=soup.find_all('table', class_='wikitable sortable sticky-header-multi sort-under')
table
```

```python
world_titles=soup.find_all('th')
```

```python
world_table_titles=[title.text.strip() for title in world_titles ]
print(world_table_titles)
```

```python
import re
```

```python
first_rank_index = world_table_titles.index('Rank')
second_rank_index = world_table_titles.index('Rank', first_rank_index + 1)
print(first_rank_index)
print(second_rank_index)
# Slice the list from the 1st 'Rank' onwards
world_table_titles = world_table_titles[first_rank_index+1:second_rank_index]

print(world_table_titles)
```

```python

world_table_titles = [title for title in world_table_titles if not re.search(r'\d', title)]

print(world_table_titles)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=world_table_titles)
df
```

```python
column_data=soup.find_all('tr')[2:50]
```

```python
column_data
```

```python
for row in column_data[2:52]:
    row_data =row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)

    length=len(df)
    df.loc[length]=individual_row_data
    
```
